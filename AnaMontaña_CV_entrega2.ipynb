```python
import pandas as pd
data = pd.read_csv('Cars93.csv')
data.keys()
```

```python
#... verificar que. no eistan datos vacios o inf 

data.isnull().sum()
```

```python
#.. seleccionar x, y el. objetivo. Y
keys=['Cylinders','EngineSize', 'Horsepower', 'RPM', 'Fuel.tank.capacity']
y =['Price']
```

```python
# visualizar datos

import matplotlib.pyplot as plt

fig=plt.figure(figsize=(7,5))
plt.hist(data['Cylinders'])
plt.xlabel('Cy')
plt.ylabel('Counts')
plt.show()
```

```python
# visualizar datos

import matplotlib.pyplot as plt

fig=plt.figure(figsize=(7,5))
plt.hist(data['Price'])
plt.xlabel('Price')
plt.ylabel('Counts')
plt.show()
```

```python
for k in keys:
  fig=plt.figure(figsize=(7,5))
  plt.plot(data[k],data[y],'*')
  plt.xlabel(k)
  plt.ylabel('Price')
  plt.show()
```

```python
#poner el objetivo como binario

import numpy as np

Y = np.array(data[y])
print(np.unique(Y, return_counts=True))

Y[Y<25] = 0
Y[Y>=25] =1 #caro
print(np.unique(Y, return_counts=True))
```

```python
#visualizar objetivo con nuevo Y
fig=plt.figure(figsize=(7,5))
plt.hist(Y)
plt.xlabel('Price')
plt.ylabel('Counts')
plt.show()
```

```python
#clasificacion visual

for k in keys:
  fig=plt.figure(figsize=(7,5))
  plt.plot(data[k],Y,'*')
  plt.xlabel(k)
  plt.ylabel('quality')
  plt.show()
```
